```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define a quantidade de dados
N = 20
# Cria vetor de N posições (0 a N)
x = np.arange(N)
# Cria vetor de N elementos gerados aleatóriamente
y = np.random.random(N)
```

```python
# Primeira visualização: Plot simples
plt.plot(x,y)
plt.show()
```

```python
# Alterando algumas propriedades da visualização:
# Tamanho da figura
plt.figure(figsize=(20,10))
# Título
plt.title('Figura 1. Plot simples')
# Quadriculado
plt.grid()
# Rótulo nos eixos
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
# Ticks de rótulo
plt.xticks(x)
# Dados e marcador (vai mudando pra ver a diferença) 
plt.plot(x,y)
#plt.plot(x,y, marker='x', color='green', ls='--')
#plt.plot(x,y, marker='.', color='red', ls='-')
#plt.plot(x,y, marker=',', color='blue', ls=':')
#plt.plot(x,y, marker='+', color='orange')
plt.show()
```
